### 도전문제
- 10000 보다 작은 소수를 모두 구하시오.

```python
import math

#math.sqrt() : 제곱근 적용

소수 = [2]

for n in range(3,10000,2):
    found = False
    for i in 소수:
        if i>math.sqrt(n): break
            
        if n%i==0:
            found=True
            break
    if found==False:
        소수.append(n)
        
소수
```
