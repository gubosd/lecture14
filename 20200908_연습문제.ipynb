연습문제
======

iris 데이터를 불러오자.

1. iris 의 50번째~99번째 까지의 샘플에서
두번째,세번째,다섯번째 속성을 뽑아내시오.
(2차원 리스트로 생성)

2. 각 속성의 평균을 구하고,
각 속성에서 해당 평균을 뺀 결과 리스트를 구하시오.

> 깃허브의 어제 강의노트 참고

```python
f = open('iris.csv')

s = f.readline()
header = s.strip().split(',')

labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
iris = []

for line in f:
    l = line.strip().split(',')
    l[0] = float(l[0])
    l[1] = float(l[1])
    l[2] = float(l[2])
    l[3] = float(l[3])
    l[4] = labels.index(l[4])
    iris.append(l)

f.close()

iris
```

```python
data = iris[50:100]
data
```

```python
[[i[1],i[2],i[-1]] for i in data]
```

```python
[[i[1],i[2],i[-1]] for i in iris[50:100]]
```

```python
m = [0,0,0,0]

for i in iris:
    for j in range(4):
        m[j] += i[j]
    
m = [i/150 for i in m]
m
```

```python
avg = [0,0,0,0]

for i in iris:
    for j in range(4):
        avg[j] += i[j]
avg = [i/len(iris) for i in avg]

# 빼기
iris2 = iris.copy()

for i in iris2:
    for j in range(4):
        i[j] -= avg[j]
iris2
```

```python
avg2 = [0,0,0,0]

for i in iris2:
    for j in range(4):
        avg2[j] += i[j]
avg2 = [i/len(iris2) for i in avg2]
avg2
```
