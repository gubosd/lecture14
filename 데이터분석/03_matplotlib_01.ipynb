### 시각화의 중요성과 matplotlib
- 인간은 가장 시각 능력이 뛰어난 동물 중의 하나이다.
- 하지만 인간의 수치 계산 능력은 본능이 아니다. (언어의 발달로 인한 논리 능력의 확장임)
- 이런 점들이 인간이 아직까지 인공지능에 비해 시각적 판단이 뛰어난 이유이다.
- 시각 능력은 다른 말로 하면 **패턴인식** 능력이다. 빛의 분포로 생기는 시각 패턴을 인간은 너무나 쉽게 알아 볼 수 있다.
- 이러한 인간의 능력을 최대한 발휘하기 위해서 숫자로 되어 있는 데이터를 시각화 하여야 한다.
- 제공받은 데이터의 **특성을 분석**하기 위해서 뿐만 아니라, **결과를 설명**하기 위해서 시각화는 아주 중요하다.
- matplotlib 는 파이썬의 시각화 도구이다.
- matplotlib 는 MATLAB 의 시각화 기능을 참고하여 만들어졌다.
- matplotlib 는 3차원 시각화 도구를 제공하지만 평면 상에서 3차원 도표는 알아보기 쉽지 않으므로, 대부분의 시각화 기능은 2차원을 위주로 한다.

### 등고선
- 이차원 평면 상에서 등고선을 그린다.

```python
x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)

X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2

plt.imshow(Z, origin='lower')
plt.colorbar()
```

```python
x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)

X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2

CS = plt.contour(X, Y, Z, levels=np.arange(-1,1,0.2)) # plt.contour(zz)
plt.clabel(CS, inline=2, fontsize=10)
plt.colorbar()
```

```python
x = np.arange(-2,2,0.1)
y = np.arange(-2,2,0.1)

X, Y = np.meshgrid(x,y)
Z = np.exp(-X**2-Y**2)

plt.contourf(X, Y, Z, levels=[0, 0.4, 0.7, 0.9, 0.97, 0.99, 1],alpha=0.5)
plt.colorbar()
CS = plt.contour(X, Y, Z, levels=[0, 0.4, 0.7, 0.9, 0.97, 0.99, 1])
plt.clabel(CS, inline=2, fontsize=10)
```

### 여러개의 플롯을 한번에 그리기

```python
fig, axes = plt.subplots(2,2, figsize=[10,8])
display(axes.shape)

axes[0,0].plot([1,2,3,4])
axes[0,1].scatter(np.random.randn(100), np.random.randn(100))
axes[1,0].bar(range(10),range(10))
axes[1,1].imshow(np.random.randn(64,64))

fig.suptitle('TEST', fontsize=20)
axes[0,0].set_title('TEST1')
```

### 플롯을 파일로 저장

```python
x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)

xx, yy = np.meshgrid(x,y)

a = np.sin(xx*yy)
plt.imshow(a, origin='lower')
plt.colorbar()
plt.xticks(range(20),x, rotation=90)

plt.savefig('test.png', dpi=50)
```

### 3차원 그림 맛보기

```python
from mpl_toolkits.mplot3d import Axes3D, axes3d

fig = plt.figure()
#ax=Axes3D(fig, elev=-152, azim=-26)
ax = Axes3D(fig)

rng = np.arange(-10,10,0.5)
X,Y = np.meshgrid(rng, rng)
#Z = X**2 + Y**2
Z = 1/(1+np.exp(X-Y))

m = ax.scatter(X,Y,Z, s=5, cmap='autumn')
#fig.colorbar(m, shrink=0.7)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

### 확률밀도함수

```python
import numpy as np
import seaborn as sns

data = [1.5]*7 + [2.5]*2 + [3.5]*8 + [4.5]*3 + [5.5]*1 + [6.5]*8
plt.subplot(2,1,1)
plt.hist(data)
plt.xlim(0,8)

plt.subplot(2,1,2)
sns.set_style('whitegrid')
sns.kdeplot(np.array(data), bw=0.5)
plt.xlim(0,8)
```
