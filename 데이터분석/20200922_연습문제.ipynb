- 위의 데이터에 어제 배운 pandas의 세 종류 함수를 적용해 보자.
> - apply(), applymap(), map()
> - value_counts()
> - sort_values()

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
whr = pd.read_csv('WHR20_DataForFigure2.1.csv')
whr
```

```python
print(*whr.columns, sep='\n')
```

```python
whr['Regional indicator'].unique()
```

```python
whr['Regional indicator'].value_counts()
```

```python
whr[whr['Regional indicator']=='East Asia']
```

```python
whr.sort_values(['Regional indicator', 'Country name'])
```

```python
whr['Logged GDP per capita'].map(np.exp)
```

```python
plt.hist(whr['Generosity'])
```

```python
whr.groupby('Regional indicator').mean()[['Ladder score','Logged GDP per capita']]
```

```python
whr.groupby('Regional indicator').agg(['count','mean'])[['Ladder score','Logged GDP per capita']]
```

```python
whr[whr['Regional indicator']=='North America and ANZ']
```

```python
whr.groupby('Regional indicator').apply(lambda df: df.sort_values('Ladder score').iloc[[-1,0]])
```

```python
plt.scatter(np.exp(whr['Ladder score']), whr['Generosity'])
```

```python
whr.sort_values('Generosity')[-5:]
```
