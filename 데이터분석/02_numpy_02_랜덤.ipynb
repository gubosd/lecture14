# Numpy 랜덤 생성 함수
- np.random.rand() : 0~1 사이의 실수값
- np.random.randn() : 평균이 0, 표준편차가 1인 정규분포에서 실수값을 뽑는다
- np.random.randint() : 정해진 범위에서 정수값을 뽑는다
- 
- np.random.normal() : 정규분포에서 평균과 표준편차를 직접 지정
- np.random.uniform() : 정해진 범위에서 실수값을 뽑는다
- 
- np.random.permutation() : 데이터를 임의로 섞어 새로운 array 를 생성한다
- np.random.choice() : 데이터에서 정해진 갯수만큼 값을 뽑아낸다
- np.random.seed() : 실험 재현을 위해 시드값을 고정한다

### 연습문제
- 10x10 의 격자가 있다.
- (0,0) 에서 출발하여 (9,9) 위치로 가면 성공이다.
- 지나간 경로를 plt.imshow() 함수로 그려보시오.
- 단, 왼/오/상/하 로 이동할 확률은 동일하다.
- 몇 번 만에 성공하는지도 확인해 보시오.

```python
board = np.zeros([10,10])
x = 0
y = 0
board[x,y] = 1

while True:
    action = np.random.randint(4)
    if action==0: # left
        if x>0:
            x = x-1
    elif action==1: # right
        if x<9:
            x = x+1
    elif action==2: # up
        if y>0:
            y = y-1
    elif action==3: # down
        if y<9:
            y = y+1
            
    print(x,y,action)
    board[x,y] +=1
    
    if x==9 and y==9:
        break
        
plt.imshow(board, cmap='gray_r')
```
