### 연습문제
- 아래는 학생들의 성적표이다.
- 각 과목의 가중치 W 를 주어 학생별 평균을 구하려고 한다.
- 문제1) numpy 의 곱셈을 이용하는 방법
- 문제2) 행렬곱을 이용하는 방법

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
score = np.random.randint(50,100,size=[100,3])
W = np.array([1, 0.5, 0.7])

display(score, W)
```

```python
(score*W).mean(axis=1)
```

```python
(score@W)/3
```

```python
W = np.array([[1,0.5],[0.5,0.9],[0.7,0.6]])
W
```

```python
score@W
```
