### world happiness report 2020 (csv)

```python
whr = pd.read_csv('WHR20_DataForFigure2.1.csv')
whr
```

```python
plt.scatter(whr['Logged GDP per capita'], whr['Ladder score'])
plt.scatter(whr['Logged GDP per capita'][60:61], whr['Ladder score'][60:61])
plt.xlabel('GDP/person(log scale)')
plt.ylabel('Happiness')
```

```python
plt.scatter(np.exp(whr['Logged GDP per capita']), whr['Ladder score'])
plt.scatter(np.exp(whr['Logged GDP per capita'][60:61]), whr['Ladder score'][60:61])
plt.xlabel('GDP/person')
plt.ylabel('Happiness')
```

```python
plt.scatter(whr['Social support'], whr['Ladder score'])
plt.scatter(whr['Social support'][60:61], whr['Ladder score'][60:61])
plt.xlabel('GDP/person')
plt.ylabel('Happiness')
```

```python
whr[(whr['Ladder score']>5.87) & (whr['Logged GDP per capita']<10.51)].sort_values('Logged GDP per capita')
```
