### 연습문제
- 동전을 100번 던지는 것을 한 게임이라고 하자.
> 앞면이 나오면 이기는 것이다.
- 이런 게임을 10000번 했을 때, 이를 numpy 어레이로 표현해 보자.
> np.random.randint() 함수 사용
- 각 게임에서 이긴 횟수를 기록하고, 그 결과를 아래와 같은 리스트로 나타내자.
> [0번 이긴 게임수, 1번 이긴 게임수, ... , 100번 이긴 게임수]
- 위의 결과를 plt.plot() 함수로 그려보자.
> 종모양 곡선이 나오는지 확인해 보시오.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
match = np.random.randint(2, size=[10000,100])
match
```

```python
match.shape
```

```python
wins = [game.sum() for game in match] # wins = match.sum(axis=1)
wins
```

```python
counts = np.zeros(101)
for i in wins:
    counts[i] += 1 # 이긴 게임수에 해당하는 카운트를 증가시킴
    
counts

# counts = np.bincount(wins)
```

```python
plt.plot(counts)
```
