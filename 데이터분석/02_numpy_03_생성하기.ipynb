### 그래프 맛보기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.imshow([[0,1,0], [2,0,3], [0,4,0]], cmap='gray')
plt.colorbar()
```

```python
a = np.arange(256).reshape(16,16)
a
```

```python
plt.imshow(a, cmap='gray')
plt.colorbar()
```

```python
a = np.zeros([20,20])
a[:3] = 1
a[-1,4:7] = 2
a[5, 2] = 3

plt.imshow(a, cmap='cool')
plt.colorbar()
```

```python
a = np.random.rand(480, 640)

plt.imshow(a, cmap='gray')
plt.colorbar()
```

```python
a = np.random.rand(480, 640, 3)

plt.imshow(a)
```

```python
a = np.random.randn(480, 640)

plt.imshow(a, cmap='gray')
plt.colorbar()
```

```python
(a>0.5).astype(int)
```

```python
a = np.random.rand(480, 640)

plt.imshow((a>0.5).astype(int), cmap='gray')
#plt.imshow(np.random.randint(2, size=(480,640)), cmap='gray')
plt.colorbar()
```

```python
a = np.random.choice([0,1], p=[0.999, 0.001], size=[480,640])

plt.imshow(a, cmap='gray')
plt.colorbar()
```

```python
a = np.random.normal(0, 3, size=(480, 640))

plt.imshow(a, cmap='gray', vmin=-2, vmax=2)
plt.colorbar()
```

- 경계선을 채워보자

```python
a = np.zeros([10,10])
for i in range(10):
    a[i,9-i] = i
    
plt.imshow(a, cmap='Greens')
plt.colorbar()
```

### 이미지 입출력
- 읽어오기 : plt.imread()
- 그리기 : plt.imshow()
- 흑백이미지는 2차원 => (높이, 넓이)
- 칼러이미지는 3차원 => (높이, 넓이, 채널)

```python
gray = plt.imread('george.jpg')
gray.shape
```

```python
gray.max(), gray.min()
```

```python
plt.imshow(gray, cmap='gray', vmin=0, vmax=255)
```

```python
a = plt.imread('snoopy.jpg')
print(a.shape, a.ndim)

plt.imshow(a)
plt.title('Snoopy')
```

```python
a
```

```python
a[0,0,0], a[0,0,1], a[0,0,2]
```

```python
a[0,0]
```

```python
a[:,:,0].shape # red 채널
```

```python
a[:,:,1].shape # green 채널
```

```python
a[:,:,2].shape # blue 채널
```

```python
plt.figure(figsize=[10,4])

plt.subplot(2,3,1)
plt.title('red')
plt.imshow(a[:,:,0], cmap='gray', vmin=0, vmax=255)
plt.subplot(2,3,2)
plt.title('green')
plt.imshow(a[:,:,1], cmap='gray', vmin=0, vmax=255)
plt.subplot(2,3,3)
plt.title('blue')
plt.imshow(a[:,:,2], cmap='gray', vmin=0, vmax=255)
plt.subplot(2,3,4)
plt.imshow(a)
```
